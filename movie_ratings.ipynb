In this assignment we will import and analyze a dataset of movie reviews. We will use pandas integrated aggregation functions to determine the average ratings by movie and by reviewer. 

```python
import pandas as pd
import numpy as np
```

Read the 'movies.csv' file into pandas:

```python
movies = pd.read_csv("C:/Users/cscam/movies.csv")
movies
```

Next, we will set the 'Reviewer' column as the index:

```python
movies = movies.set_index('Reviewer')
```

Now we can create a new column to display the average rating of each reviewer:

```python
movies['Average Rating'] = movies.mean(axis=1)
movies
```

We can transpose the index and columns:

```python
movies_transposed = movies.T
movies_transposed
```

And now we are able to add a column that shows the average rating of each movie, as well as the average rating across both movies and reviewers.

```python
movies_transposed['Average Rating'] = movies_transposed.mean(axis=1)
movies_transposed
```

Finally, we have a dataset showing our list of movies, ratings by each reviewer, average rating by reviewer, and average rating by movie. 

In some cases, it may be beneficial to normalize data before performing analysis. This process scales all numeric variables within a range and can skew values if there are outliers in the dataset, such as the weather dataset encountered in a previous assignment.

Our movie reviews dataset contains values of a common scale and is easily analyzed and understood without additional processing.
